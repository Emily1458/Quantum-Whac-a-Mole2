```python
from qiskit import QuantumCircuit
from qiskit_aer import AerSimulator
```

```python
def rgbtohex(r,g,b):
    return f'#{r:02x}{g:02x}{b:02x}'
```

```python
#number of successful whacks
whacks = 0
#number of the hole whacked, no longer needed
hole = 0
qc = QuantumCircuit(30, 30)

#for cx gate
targetsList = []
controlsList = []

gate = -1
isControl = True

#screen elements
screen = Tk()
screen.config(bg="#62C370", width=500, height=520)
screen.resizable(0, 0)
screen.attributes('-topmost', True)
tip = Label(bg='ghost white', fg='grey31', text="Play with mouse for best experience")
tip.place(x=0, y=497)
lab = Label(text="WHAC-A-MOLE!", font=("Algerian", 30), bg='#3DA5D9', fg='#fff')
lab.place(x=110, y=0)
scorelabel = Label(text=whacks, width=7, font=('Bahnschrift', 30), bg='#FEC601')
scorelabel.place(x=170, y=52)

timeLabel = Label(text=20,
                  width=2, 
                  height=1, 
                  font=('Bahnschrift', 30), 
                  bg="ghost white")
timeLabel.place(x=424, y=90)

remark = Label(text="", width=17, bg='#EA7317', fg='#fff', font="Centaur 20")
remark.place(x=130, y=107)
```

```python
#side buttons and their methods
def applyH():
    global gate
    gate = 0

def applyX():
    global gate
    gate = 1

def applyCX():
    global gate
    gate = 2

def applyZ():
    global gate
    gate = 3

hGate = Button(
    width=6,
    height=3,
    bg='ghost white',
    text='h-gate',
    state='normal',
    command=applyH)
hGate.place(x = 421, y =160)

xGate = Button(
    width=6,
    height=3,
    bg='ghost white',
    text='x-gate',
    state='normal',
    command=applyX)
xGate.place(x = 421, y =225)

cnotGate = Button(
    width=6,
    height=3,
    bg='ghost white',
    text='CNOT',
    state='normal',
    command=applyCX)
cnotGate.place(x = 421, y =290)

zGate = Button(
    width=6,
    height=3,
    bg='ghost white',
    text='z-gate',
    state='normal',
    command=applyZ)
zGate.place(x = 421, y =355)
```

```python
#runs the game for 20 seconds after start is pressed
def whac_a_mole():
    ready_set_whack()
    global hole
    for i in range(0, 20):
        sleep(1)
        timeLabel.config(text=20-i-1)
    replaybtn.config(state='normal')
    global whacks
    if whacks < 5:
        remark.config(text="POOR")
    elif whacks >= 5 and whacks < 10:
        remark.config(text="BAD")
    elif whacks >= 10 and whacks < 15:
        remark.config(text="DO BETTER!")
    elif whacks >= 15 and whacks < 20:
        remark.config(text="GOOD")
    elif whacks >= 20 and whacks < 25:
        remark.config(text="GREAT!")
    elif whacks >= 25 and whacks < 35:
        remark.config(text="AWESOME!")
    elif whacks == 35:
        remark.config(text="WHACKED 'em ALL!", font="Centaur 18")

#when start button is pressed, whac_a_mole() is run in a separate thread
def start():
    global whacks
    whacks = 0
    global gate
    gate = -1
    
    remark.config(text="")
    replaybtn.config(state='disabled')
    hGate.config(state="disabled")
    xGate.config(state="disabled")
    zGate.config(state="disabled")
    startButton.config(state="disabled")
    cnotGate.config(state="disabled")
    t = threading.Thread(target=whac_a_mole)
    t.start()
    

startButton = Button(
    width=6,
    height=3,
    bg='ghost white',
    text='start!',
    state='normal',
    command=start)
startButton.place(x=19, y=90)

def replay():
    global qc
    qc = QuantumCircuit(30, 30)
    gate = 100
    for b in buttons:
        b.config(bg="#d5a36a",
            text='【•】__【•】',
            fg="#595031", 
            state = "normal")
    hGate.config(state="normal")
    xGate.config(state="normal")
    zGate.config(state="normal")
    cnotGate.config(state="normal")
    startButton.config(state="normal")
    global gateList
    gateList = ["【•】__【•】\n"] * 30
    global controlsList
    global targetsList
    controlsList = []
    targetsList = []

#configures the replay button
replaybtn = Button(text="Play Again", command=replay)
replaybtn.place(x=0, y=2)


    
```

```python
#creates the array of buttons (moles/qubits)

xPos = [19, 86, 153, 220, 287, 354]
#421
yPos = [160, 225, 290, 355, 419]

buttons = []

for xP in xPos:
    for yP in yPos:
        buttons.append(Button(
            width=6,
            height=3,
            bg="#d5a36a",
            text='【•】__【•】',
            fg="#595031",
            state='normal',
            command=lambda m=len(buttons): onwhack(m)))
        buttons[len(buttons) - 1].place(x = xP, y = yP)
```

```python
#event triggered when mole is clicked

def onwhack(num):
    global gate
    if gate >= 0:
        placeGate(num, gate)
    else:
        global hole
        global whacks
        toMeasure = findAllEntangled(num)
        
        qc.measure(range(30), range(30))
        sim = Aer.get_backend('qasm_simulator')
        qexe = transpile(qc, sim)
        job = sim.run(qexe, shots = 1)
        result = job.result()

        for b in toMeasure:
            if quantumMeasure(b, result) == 1:
                whacks += 1
                scorelabel.config(text="* " + str(whacks) + " *")
                buttons[b].config(state='disabled', text='x_x', fg='black')
            elif quantumMeasure(b, result) == 0:
                buttons[b].config(state='disabled', text='◉‿◉', fg='black')
```

```python
#stores format of mole face
gateList = ["【•】__【•】\n"] * 30
target = -1
control = -1

#num: the number of the button
#g: same as the global variable gate
def placeGate(num, g):
    global target
    global control 
    global gateList
    global isControl 
    #H gate
    if g == 0:
        qc.h(num)
        gateList[num] = gateList[num] + "H "
        buttons[num].config(bg='#e677af', text=gateList[num])
    #X gate
    if g == 1:
        qc.x(num)
        gateList[num] = gateList[num] + "X "
        buttons[num].config(bg='#5ec2bf', text=gateList[num])
    #CX gate
    if g == 2:
        if isControl == True:
            control = num
            controlsList.append(num)
            gateList[num] = gateList[num] + "CX:C" + str(len(controlsList)) + ""
            buttons[num].config(bg='#ecbe44', text=gateList[num])
            isControl = False
        else:
            target = num
            qc.cx(control, target)
            targetsList.append(num)
            gateList[num] = gateList[num] + "CX:T" + str(len(targetsList)) + ""
            buttons[num].config(bg='#ecbe44', text=gateList[num])
            isControl = True
        #print(targetsList)
        #print(controlsList)
    #Z gate
    if g== 3:
        gateList[num] = gateList[num] + "Z "
        buttons[num].config(bg='#1e86a3', text=gateList[num])
```

```python
def placeCX(c, t): 
    pair = {
        "control": c, 
        "target": t
    }
    
```

```python
#happens when mole is clicked outside of place gate time
isLookingInT = True

def findAllEntangled(start):
    entangled = [start]
    
    if start in controlsList:
        entangled.append(start)
        i = controlsList.index(start)
        m = 0
        while m < 1:
            entangled.append(targetsList[i])
            u = entangled[len(entangled) - 1]
            if (u not in controlsList):
                break
            i = controlsList.index(u)

    if start in targetsList:
        i = targetsList.index(start)
        m = 0
        while m < 1:
            entangled.append(controlsList[i])
            u = entangled[len(entangled) - 1]
            if (u not in targetsList):
                break
            i = targetsList.index(u)
        
    return entangled

def quantumMeasure(bnum, results):
    
    counts = results.get_counts()
    numberList = list(counts)
    number = numberList[0]
    bit = number[30 - bnum - 1]
    return int(bit)
    #buttons[bnum].config(bg=rgbtohex(255, 255*col, 255))
    
```

```python
#the message banner changes after start button is clicked
def ready_set_whack():
    for b in buttons: 
        b.config(state='normal')
```
